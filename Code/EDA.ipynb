```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
```

```python
df = pd.read_csv('../data/all_clean.csv')
df
```

```python
def label_hist_plot(df, columns):
    ax = sns.countplot(data = df, x=columns)
    ax.set_xticklabels(["Down",'Up'])
    ax.set(ylabel='Count of label')
    plt.show()
    # plt.legend(title='SP 500 Mean', labels=["Up",'Down'])
label_hist_plot(df,df.sp500_mean_label)
label_hist_plot(df,df.sp500_last_label)
label_hist_plot(df,df.usdx_mean_label)
label_hist_plot(df,df.usdx_last_label)
```

```python
# Wordcloud to display words frequency after cleaning data
from wordcloud import WordCloud, STOPWORDS
text = " ".join(i for i in df.text)
STOPWORDS.update(['t', 'co', 'tweet', 'amp'])
stopwords = set(STOPWORDS)
wordcloud = WordCloud(stopwords = stopwords, collocations=False, width = 800, height = 400, background_color="white").generate(text)
plt.figure(figsize=(10,5), facecolor = None)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.tight_layout(pad = 0)
plt.show()
```
