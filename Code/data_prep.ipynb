# Prepare for Naive Bayes Model

```python
### IMPORT PACKAGES 
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
import numpy as np
import pandas as pd
```

```python
# Read Data
df = pd.read_csv('../data/all_clean.csv')
corpus = df.text
```

### Plain CountVectorizer

```python
# INITIALIZE COUNT VECTORIZER
vectorizer=CountVectorizer(min_df=10,stop_words='english')
# RUN COUNT VECTORIZER ON OUR COURPUS 
Xs  =  vectorizer.fit_transform(corpus)
#CONVERT TO DENSE MATRIX
X= np.array(Xs.todense())
df_X = pd.DataFrame(X)
df_X
```

```python
out = pd.merge(df.drop(columns = ['text','Unnamed: 0']), df_X, left_index=True, right_index=True)
out.to_csv('../data/matrix_CV_plain.csv')
```

### Ngram CountVectorizer

```python
# INITIALIZE COUNT VECTORIZER
vectorizer=CountVectorizer(ngram_range=(1,4),min_df=10,stop_words='english')
# RUN COUNT VECTORIZER ON OUR COURPUS 
Xs  =  vectorizer.fit_transform(corpus)
#CONVERT TO DENSE MATRIX
X= np.array(Xs.todense())
df_X = pd.DataFrame(X)
df_X
```

```python
out = pd.merge(df.drop(columns = ['text','Unnamed: 0']), df_X, left_index=True, right_index=True)
out.to_csv('../data/matrix_CV_ngram.csv')
```

### Ngram TfidfVectorizer

```python

vectorizer=TfidfVectorizer(ngram_range=(1,4),min_df=10,stop_words='english')
# RUN TFIDF VECTORIZER ON OUR COURPUS 
X_ngram  =  vectorizer.fit_transform(corpus)
#CONVERT TO DENSE MATRIX
X_ngram= np.array(X_ngram.todense())
df_X = pd.DataFrame(X_ngram)
df_X
```

```python
out = pd.merge(df.drop(columns = ['text','Unnamed: 0']), df_X, left_index=True, right_index=True)
out.to_csv('../data/matrix_tfidf_ngram.csv')
```
